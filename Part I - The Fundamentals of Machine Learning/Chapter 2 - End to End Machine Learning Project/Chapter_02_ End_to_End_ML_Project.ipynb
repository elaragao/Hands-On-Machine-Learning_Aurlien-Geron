## Select a Performance Measure <a id="section_2_2"></a>

It is necessary to choose a Performance Measure, which measures how much the algorithm misses its prediction. Although there are multiple metrics for this, we will look at two:


\begin{equation}
RMSE(\mathbf{X},h) = \sqrt{
\frac{1}{m}
\sum_{i=1}^{m}
(h(x^{(i)})-y^{(i)})^{2}
}\end{equation}



Although RMSE is the most widely used measure because it highlights outliers, the Mean Absolute Error (MAE) is also used.

\begin{equation}
MAE(\mathbf{X},h) = 
\frac{1}{m}
\sum_{i=1}^{m}
|
h(x^{(i)})-y^{(i)}
|
\end{equation}

### Notations

- **h** is the system’s prediction function, also called a _hypothesis_.
- **X** is a matrix containing all the feature values (excluding labels) of all dataset.
- **m** is the number of instances in the dataset.
- **x** is a vector of all the feature values (excluding the label) of the i<sup>th</sup> instance.
- **y** is the labels of the dataset of the i<sup>th</sup> instance.


### Norms

The most common norms are the Manhattan and Euclidean norms, which are ways of measuring distances between points in a space, widely used in mathematics, statistics and machine learning.

#### Manhattan norm
Also called L1 distance or taxicab norm, it measures the distance traveled along an orthogonal path (or in straight and perpendicular lines), such as a taxi moving on city streets. Mathematically, it is the sum of the absolute differences between the coordinates of the points:

\begin{equation}
L1 = \sum^{n}_{i=1} |x_{i} - y_{i}| 
\end{equation}


Applications: Problems where movement is restricted to orthogonal axes, such as board games, traffic networks or some metrics in machine learning.

#### Euclidean Norm
Also called L2 distance, it is the measurement of the straight line between two points in a space. It is based on the Pythagorean theorem and calculates the square root of the sum of the squares of the differences between the coordinates:

\begin{equation}
L2 = \sqrt{\sum^{n}_{i=1} (x_{i} - y_{i})^{2}}
\end{equation}

Applications: Geometric models, machine learning (such as k-NN), and data analysis where Euclidean space is appropriate.

## Download the Data <a id="section_3_5"></a>

```python
from pathlib import Path
import pandas as pd
import tarfile
import urllib.request

def load_housing_data():
    tarball_path = Path("datasets/housing.tgz")
    if not tarball_path.is_file(): # If there is no such file in dataset folder
        Path("datasets").mkdir(parents=True, exist_ok=True) # Create path if it not exist
        url = "https://github.com/ageron/data/raw/main/housing.tgz" # Search URL
        urllib.request.urlretrieve(url, tarball_path) # Request download from url
        with tarfile.open(tarball_path) as housing_tarball: # Extract files in folder
            housing_tarball.extractall(path="datasets")
    return pd.read_csv(Path("datasets/housing/housing.csv")) # Read extracted file as pd

housing = load_housing_data()
```

## Create a Test Set <a id="section_3_7"></a>

The first step is to split the data into training and testing sets:

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)
```

```python
test_set["total_bedrooms"].isnull().sum()
```

Having a large enough dataset would be enough, but since it is not, it is necessary to take care of the sampling bias.

To do this, <i>stratified sampling</i> are necessary, which is to divide the population into homogeneous subgroups called <i>strata</i>, to ensure that the set is representative of the entire population.

What will be observed as an important attribute is the `median_income`. It is known that most of the income in this data set varies between \\$15,000 and \\$60,000. So it is useful to use the `pd.cut()` function to create a categorical attribute from a numeric attribute:

```python
import numpy as np

housing["income_cat"] = pd.cut(housing["median_income"],
                               bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                               labels=[1, 2, 3, 4, 5])
```

```python
housing["income_cat"].value_counts()/len(housing["income_cat"])
```

Attribute 1 will be represented by values ​​between 0 and 1.5, attribute 2 will be represented by values between 1.5 and 3, attribute 3 between 3 and 4.5, 4 between 4.5 and 6, and 5 will be between 6 and the highest value it has in the dataset. It is also useful to visualize how the distribution behaves in a histogram:

```python
import matplotlib.pyplot as plt

housing["income_cat"].value_counts().sort_index().plot.bar(rot=0, grid=True)
plt.xlabel("Income category")
plt.ylabel("Number of districts")
save_fig("housing_income_cat_bar_plot")  # extra code
plt.show()
```

Using the `stratify` argument, it is possible to do this separation with just one line:

```python
strat_train_set, strat_test_set = train_test_split(
    housing, test_size=0.2, stratify=housing["income_cat"], random_state=42)
```

```python
strat_test_set["income_cat"].value_counts() / len(strat_test_set)
```

```python
# Extra code to observe the comparison of the median_income distributions of the total and stratified values


def income_cat_proportions(data):
    return data["income_cat"].value_counts() / len(data)

train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)

compare_props = pd.DataFrame({
    "Overall %": income_cat_proportions(housing),
    "Stratified %": income_cat_proportions(strat_test_set),
    "Random %": income_cat_proportions(test_set),
}).sort_index()
compare_props.index.name = "Income Category"
compare_props["Strat. Error %"] = (compare_props["Stratified %"] /
                                   compare_props["Overall %"] - 1)
compare_props["Rand. Error %"] = (compare_props["Random %"] /
                                  compare_props["Overall %"] - 1)
(compare_props * 100).round(2)

# It is possible to observe the difference between a stratified division and a normal division, with the normal division being 
# something with much more error
```

```python
print("Strat Error %")
print(compare_props['Strat. Error %'].abs().sum()*100)

print("Rand. Error %")
print(compare_props['Rand. Error %'].abs().sum()*100)
```

```python
# The feature `income_cat` will not be used again, so we'll just drop it

for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis=1, inplace=True)
```
