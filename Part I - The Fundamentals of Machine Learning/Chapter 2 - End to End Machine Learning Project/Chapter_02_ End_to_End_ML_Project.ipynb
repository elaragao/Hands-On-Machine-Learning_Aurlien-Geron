## Select a Performance Measure <a id="section_2_2"></a>

It is necessary to choose a Performance Measure, which measures how much the algorithm misses its prediction. Although there are multiple metrics for this, we will look at two:


\begin{equation}
RMSE(\mathbf{X},h) = \sqrt{
\frac{1}{m}
\sum_{i=1}^{m}
(h(x^{(i)})-y^{(i)})^{2}
}\end{equation}



Although RMSE is the most widely used measure because it highlights outliers, the Mean Absolute Error (MAE) is also used.

\begin{equation}
MAE(\mathbf{X},h) = 
\frac{1}{m}
\sum_{i=1}^{m}
|
h(x^{(i)})-y^{(i)}
|
\end{equation}

### Notations

- **h** is the systemâ€™s prediction function, also called a _hypothesis_.
- **X** is a matrix containing all the feature values (excluding labels) of all dataset.
- **m** is the number of instances in the dataset.
- **x** is a vector of all the feature values (excluding the label) of the i<sup>th</sup> instance.
- **y** is the labels of the dataset of the i<sup>th</sup> instance.


### Norms

The most common norms are the Manhattan and Euclidean norms, which are ways of measuring distances between points in a space, widely used in mathematics, statistics and machine learning.

#### Manhattan norm
Also called L1 distance or taxicab norm, it measures the distance traveled along an orthogonal path (or in straight and perpendicular lines), such as a taxi moving on city streets. Mathematically, it is the sum of the absolute differences between the coordinates of the points:

\begin{equation}
L1 = \sum^{n}_{i=1} |x_{i} - y_{i}| 
\end{equation}


Applications: Problems where movement is restricted to orthogonal axes, such as board games, traffic networks or some metrics in machine learning.

#### Euclidean Norm
Also called L2 distance, it is the measurement of the straight line between two points in a space. It is based on the Pythagorean theorem and calculates the square root of the sum of the squares of the differences between the coordinates:

\begin{equation}
L2 = \sqrt{\sum^{n}_{i=1} (x_{i} - y_{i})^{2}}
\end{equation}

Applications: Geometric models, machine learning (such as k-NN), and data analysis where Euclidean space is appropriate.

Having a large enough dataset would be enough, but since it is not, it is necessary to take care of the sampling bias.

To do this, <i>stratified sampling</i> are necessary, which is to divide the population into homogeneous subgroups called <i>strata</i>, to ensure that the set is representative of the entire population.

What will be observed as an important attribute is the `median_income`. It is known that most of the income in this data set varies between \\$15,000 and \\$60,000. So it is useful to use the `pd.cut()` function to create a categorical attribute from a numeric attribute:

```python
import numpy as np

housing["income_cat"] = pd.cut(housing["median_income"],
                               bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                               labels=[1, 2, 3, 4, 5])
housing["income_cat"].value_counts()/len(housing["income_cat"])
```

Attribute 1 will be represented by values â€‹â€‹between 0 and 1.5, attribute 2 will be represented by values between 1.5 and 3, attribute 3 between 3 and 4.5, 4 between 4.5 and 6, and 5 will be between 6 and the highest value it has in the dataset. It is also useful to visualize how the distribution behaves in a histogram:

```python
import matplotlib.pyplot as plt

housing["income_cat"].value_counts().sort_index().plot.bar(rot=0, grid=True)
plt.xlabel("Income category")
plt.ylabel("Number of districts")
save_fig("housing_income_cat_bar_plot")  # extra code
plt.show()
```

Using the `stratify` argument, it is possible to do this separation with just one line:

```python
strat_train_set, strat_test_set = train_test_split(
    housing, test_size=0.2, stratify=housing["income_cat"], random_state=42)
```

```python
strat_test_set["income_cat"].value_counts() / len(strat_test_set)
```

```python
# Extra code to observe the comparison of the median_income distributions of the total and stratified values


def income_cat_proportions(data):
    return data["income_cat"].value_counts() / len(data)

train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)

compare_props = pd.DataFrame({
    "Overall %": income_cat_proportions(housing),
    "Stratified %": income_cat_proportions(strat_test_set),
    "Random %": income_cat_proportions(test_set),
}).sort_index()
compare_props.index.name = "Income Category"
compare_props["Strat. Error %"] = (compare_props["Stratified %"] /
                                   compare_props["Overall %"] - 1)
compare_props["Rand. Error %"] = (compare_props["Random %"] /
                                  compare_props["Overall %"] - 1)
(compare_props * 100).round(2)

# It is possible to observe the difference between a stratified division and a normal division, with the normal division being 
# something with much more error
```

```python
print("Strat Error %")
print(compare_props['Strat. Error %'].abs().sum()*100)

print("Rand. Error %")
print(compare_props['Rand. Error %'].abs().sum()*100)
```

## Clean the Data <a id="section_5_1"></a>

There are three common ways to deal with missing features:

1 - Get rid of the instances where the problem exists

2 - Get rid of the attribute where the problem exists

3 - Replace the missing value with another (zero, mean, median, etc.). The process is called `imputation`.

Of these three options, the third is the least destructive and is the preferable. The most manual way to do this is represented by the code below:

```python
housing_option3 = housing.copy()

median = housing["total_bedrooms"].median()
housing_option3["total_bedrooms"].fillna(median, inplace=True)  # option 3
```

### SciKit-Lear Design


#### Consistency
All objects share a consistent and simple interface:

##### Estimators
Estimators that estimate parameters based on the dataset. The estimation is done by the `fit()` method, using one parameter (the dataset) or two (the labels). Other parameters are considered <i>Hyperparameters</i>. An example is the `SimpleImputer` itself.

##### Transformers
Estimators that return the transformed dataset. Uses the `transform()` method with the dataset to be transformed as a parameter. Can also use `fit_transform()` as a convenience.

##### Predictors
Estimators capable of making predictions on the dataset. Uses the `predict()` method, when taking a dataset of new instances and returning the corresponding predictions. Also has `score()` methods to measure the quality of the prediction.


#### Inspection
All hyperparameters can be accessed through variables (e.g. `imputer.strategy`) and parameters learned from the estimator can be accessed through underscored variables (e.g. `imputer.statistics_`).

#### Nonproliferation of classes
Datasets represented by arrays (NumPy) or matrices (SciPy), and hyperparameters as strings or numbers.

#### Composition
Building blocks are reusable, for example, creating Pipelines.

#### Sensible defaults
Scikit-Learn's default values â€‹â€‹are applicable for most uses.

## Feature Scaling and Transformation <a id="section_5_3"></a>

It is important that <b>BEFORE</b> standardizations are made, problems such as <i>heavy tail</i> are neutralized, leaving the distribution centralized.

One way to do this is to apply the square root to factors where there are no negative values, or in the case of a very long <i>heavy tail</i>, the logarithmic scale (<i>power law distribution</i>) in the attributes with this factor present:

```python
fig, axs = plt.subplots(1, 2, figsize=(8, 3), sharey=True)
housing["population"].hist(ax=axs[0], bins=50)
housing["population"].apply(np.log).hist(ax=axs[1], bins=50)
axs[0].set_xlabel("Population")
axs[1].set_xlabel("Log of population")
axs[0].set_ylabel("Number of districts")
save_fig("long_tail_plot")
plt.show()
```

Another approach to dealing with heavy-tailed features is to <b>buckterize</b> the feature. This means cutting its distribution into groups of approximately equal size and replacing each feature value with the index of the group to which it belongs, or you can also replace the value with its percentile, thus achieving a uniform distribution.

```python
percentiles = [np.percentile(housing["median_income"], p)
               for p in range(1, 100)]
flattened_median_income = pd.cut(housing["median_income"],
                                 bins=[-np.inf] + percentiles + [np.inf],
                                 labels=range(1, 100 + 1))
flattened_median_income.hist(bins=50)
plt.xlabel("Median income percentile")
plt.ylabel("Number of districts")
plt.show()
```

In the case of multimodal distributions (more than one explicit peak), it is possible to perform a multimodal transformation on the main peaks by adding features representing the similarity between the peaks and the feature. Usually, a Radial Basis Function (RBF) is used, where in Gaussian the value decays exponentially as it moves away from the peak, through the equation $e^{-\gamma (x - peak\_value)^{2}}$. Using the `housing_median_age` peak of 35 years

```python
from sklearn.metrics.pairwise import rbf_kernel

age_simil_35 = rbf_kernel(housing[["housing_median_age"]], [[35]], gamma=0.1)
```

```python
ages = np.linspace(housing["housing_median_age"].min(),
                   housing["housing_median_age"].max(),
                   500).reshape(-1, 1)
gamma1 = 0.1
gamma2 = 0.03
rbf1 = rbf_kernel(ages, [[35]], gamma=gamma1)
rbf2 = rbf_kernel(ages, [[35]], gamma=gamma2)

fig, ax1 = plt.subplots()

ax1.set_xlabel("Housing median age")
ax1.set_ylabel("Number of districts")
ax1.hist(housing["housing_median_age"], bins=50)

ax2 = ax1.twinx()  # create a twin axis that shares the same x-axis
color = "blue"
ax2.plot(ages, rbf1, color=color, label="gamma = 0.10")
ax2.plot(ages, rbf2, color=color, label="gamma = 0.03", linestyle="--")
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylabel("Age similarity", color=color)

plt.legend(loc="upper left")
save_fig("age_similarity_plot")
plt.show()
```

## Custom Transformers <a id="section_5_4"></a>

Although the transformations provided by Scikit-Learn are useful, it is often necessary to write functions to perform transformations, cleanings, or combinations of attributes. In cases where training is not necessary, it is possible to write functions:

<b>Log Transform + Inverse</b>

```python
from sklearn.preprocessing import FunctionTransformer

log_transformer = FunctionTransformer(np.log, inverse_func=np.exp) # Applies logarithmic function, and its inversion later when necessary
log_pop = log_transformer.transform(housing[["population"]]) # Applies transformation to the "population" feature
```

<b>RBF Gaussian</b>

```python
# RBF in "housing_median_age"
rbf_transformer = FunctionTransformer(rbf_kernel, kw_args=dict(Y=[[35.]], gamma=0.1)) # RBF applied in age 35
age_simil_35 = rbf_transformer.transform(housing[["housing_median_age"]])

age_simil_35
```

<div class="alert alert-block alert-warning">
<b>Warning:</b> rbf_kernel() does not treat the features separately. If you pass it an array with two features, it will measure the 2D distance (Euclidean) to measure similarity. e.g:
</div>

```python
# RBF in "housing_median_age"
sf_coords = 37.7749, -122.41
sf_transformer = FunctionTransformer(rbf_kernel,
                                     kw_args=dict(Y=[sf_coords], gamma=0.1))
sf_simil = sf_transformer.transform(housing[["latitude", "longitude"]])

sf_simil
```

<b>Ration Transform</b>

```python
ratio_transformer = FunctionTransformer(lambda X: X[:, [0]] / X[:, [1]])
ratio_transformer.transform(np.array([[1., 2.], [3., 4.]]))
```

## Transformation Pipelines <a id="section_5_5"></a>

```python
def column_ratio(X):
    return X[:, [0]] / X[:, [1]]

def ratio_name(function_transformer, feature_names_in):
    return ["ratio"]  # feature names out

def ratio_pipeline():
    return make_pipeline(
        SimpleImputer(strategy="median"),
        FunctionTransformer(column_ratio, feature_names_out=ratio_name),
        StandardScaler())

log_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    FunctionTransformer(np.log, feature_names_out="one-to-one"),
    StandardScaler())
cluster_simil = ClusterSimilarity(n_clusters=10, gamma=1., random_state=42)
default_num_pipeline = make_pipeline(SimpleImputer(strategy="median"),
                                     StandardScaler())
preprocessing = ColumnTransformer([
        ("bedrooms", ratio_pipeline(), ["total_bedrooms", "total_rooms"]),
        ("rooms_per_house", ratio_pipeline(), ["total_rooms", "households"]),
        ("people_per_house", ratio_pipeline(), ["population", "households"]),
        ("log", log_pipeline, ["total_bedrooms", "total_rooms", "population",
                               "households", "median_income"]),
        ("geo", cluster_simil, ["latitude", "longitude"]),
        ("cat", cat_pipeline, make_column_selector(dtype_include=object)),
    ],
    remainder=default_num_pipeline)  # one column remaining: housing_median_age
```
